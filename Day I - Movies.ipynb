# Missing Data

```python
#examining missing values
print('Missing Values Distribution:')
print(movies.isnull().mean())
print("")
```

```python
#check data type of each column
print("Column Datatypes: ")
print(movies.dtypes)
```

```python
#getting string/mixed type columns
str_cols = list(movies.columns)
str_cols.remove('RATING')
str_cols.remove('RunTime')
```

```python
#checking for trailing and leading whitespaces
for i in str_cols:
    movies[i] = movies[i].str.strip()
```

```python
#deleting the gross column
del movies['Gross']
```

```python
movies.head()
```

```python
#replacing NaN values with zero
columns = ['RATING', 'VOTES', 'RunTime']

for i in columns:
    movies[i] = movies[i].fillna(0)
```

```python
movies.head()
```

```python
#deleting rows with zero values

movies = movies[movies.RATING != 0]
```

```python
movies.head()
```

```python
movies = movies[movies.VOTES != 0]
movies.head()
```

```python
movies = movies[movies.RunTime != 0]
movies.head()
```

```python
movies = movies.reset_index(drop = True)
```
